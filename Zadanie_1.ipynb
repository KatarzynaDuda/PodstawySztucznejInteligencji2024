#Zadanie 1 - czatboty


Zadanie 1

Które z następujących zadań wymagają w Twojej opinii inteligencji od człowieka:

• wypełnianie deklaracji PIT - nie wymaga inteligencji, jest odtwórcze, można znaleść w internecie instrukcję lub wzór wypełnienia.

• streszczanie tekstu - wymaga inteligencji, trzeba umieć wyłapać istotne informacje z tekstu, zrozumieć jego sedno, zawrzeć w streszczeniu tylko najważniejsze elementy, wątki.

• tłumaczenie tekstu - wymaga inteligencji, trzeba być świadomym kontekstu, np z języka angielkiego na polski czy ,,you" znaczy ,,ty" czy ,,wy", nie tłumaczyć słowo w słowo(odtwórcze), umieć rozpoznać gdzie pojawia się fraza nieprzetłumaczalna (np. powiedzenia, przysłowia, które należy zastąpić ich odpowiednikami znaczeniowymi, które często nie używają dosłownie tych samych słów, np. ,,dziękuję z góry" nie przetłumaczymy jako ,,thank you from the mountain")

• klasyfikacja tekstu do kategorii tematycznych - nie wymaga inteligencji, po przeczytaniu paru zdań często można łatwo rozpoznać do jakiej kategorii przyporządkować tekst.

• odpowiadanie na proste pytania zadawane w języku naturalnym (np. polskim) - nie wymaga inteligencji, robimy to codziennie bez zastanowienia, np. small talk.

• układanie rozkładu jazdy transportu miejskiego - wymaga inteligencji, żeby np. nie jeździł ten sam autobus co minutę, dopasować godziny przyjazdów na przystanek do godzin kiedy ludzie zaczynają i kończą pracę lub szkołę, trzeba być świadomym, że nie o każdej godzinie trasa autobusu będzie trwała tyle samo, w godzinach sczytu zajmie to więcej czasu.

• programowanie (pisanie programów komputerowych) - wymaga inteligencji, myślenia logicznego, trzeba wymyśleć w jakiej kolejności umieszczać polecenia aby program poprawnie wykonał to co chcemy.

• „programowanie” kanałów telewizyjnych - wymaga inteligencji, ale nie bardzo dużo, trzeba wiedzieć jakie opcje klikać, wiedza techniczna, zadanie mechaniczne, odtwórcze, kiedy raz się nauczymy będziemy to robić bez problemu.

• testowanie oprogramowania - wymaga inteligencji, wymaga wiedzy programistycznej, wiedzy gdzie często pojawiają się błędy, co dokładnie testować.

• komponowanie muzyki - zależy jaki rodzaj muzyki komponujemy, są gatunki, w których często powtarza się specjalny układ akordów, jest to swego rodzaju instrukcja tworzenia melodii, nie wymaga to dużo inteligencji, z kolei komponowanie np. wieloinstrumentowych utworów wymaga inteligencji oraz wiedzy muzycznej na temat instrumentów.

• rozwiązywanie układów równań - wymaga inteligencji, trzeba znać sposoby rozwiązywania, umieć znajdować/zauważać zależności między niewiadomymi i liczbami.

• symboliczne obliczanie pochodnych funkcji - dla łatwych przypadków nie wymaga inteligencji, należy podstawić do wzoru, jednak przy bardziej skąplikowanych funkcjach inteligencja oraz wiedza matematyczna są konieczne.

• symboliczne całkowanie funkcji - wymaga inteligencji, wiedzy matematycznej.

• kierowanie samochodem - wymaga inteligencji, trzeba umieć koordynować odpowiednie ruchy np. naciskanie pedałów i zmiana biegów, trzeba być uważnym i świadomym tego co się dzieje w otoczeniu, szybko reagować na niespodziewane sytuacje.

Zadanie 2


Które z następujących problemów można uznać za mieszczące się w zakresie sztucznej inteligencji:

• streszczanie tekstu - AI potrafi streścić tekst, jednak nie zawsze robi to dobrze, może nie wyłapać informacji ważnych. Chat GPT poproszony o streszczenie tekstu stworzył treść dłuższą niż oryginalna.

• tłumaczenie tekstu - AI moze wykonać tłumaczenie dobrej jakości, jednak problemy pojawiają się z powiedzeniami, przyimkami, warto sprawdzić tłumaczenia proponowane przez AI samemu.

• klasyfikacja tekstu do kategorii tematycznych - AI może wykonać to zadanie, można zaprogramować słowa kluczowe których nalezy szukać w tekstach aby tekst należał do danej kategorii.

• odpowiadanie na proste pytania zadawane w języku naturalnym - tak, rozmowy takie są schematyczne, AI potrafi je układać.

• rozwiązywanie układów równań - tak, AI potrafi operować na liczbach, rozwiązywać równania, może podać proces, kroki rozwiązywania.

• układanie rozkładu jazdy - AI może to zrobic, obliczyć optymalną trasę, czas przejazdu autobusu, jednak należy zaopatrzyć AI w dane dotyczące wymagań do trasy, w jakich godzinach autobusy powinny jeździc częściej, jakie są ważne miejsca, w których mają się zatrzymać.

• rozwiązywanie układów równań liniowych - tak, AI potrafi wykonywać obliczenia matematyczne.

• symboliczne obliczanie pochodnych - tak.

• symboliczne całkowanie - tak.

• kierowanie samochodem - tak, jeśli zamochód jest wyposażony w bardzo dobre czujniki z dużym zasięgiem, oraz ma zaprogramowane zasady ruchu drogowego.

Zadanie 3

Które z poniższych rodzajów komunikacyjnego zachowania człowieka mogą być
obecnie skutecznie imitowane przez sztuczne systemy (odpowiednio oprogramowane maszyny):

• rozmowa towarzyska - tak, chat GPT i inne chatboty potrafią dobrze naślasdować luźne rozmowy.

• dyskusja polityczna - tak, jeśli ma dostęp do bazy danych reprezentującej poglądy danej grupy politycznej.

• dyskusja naukowa - tak, jeśli ma dobrą baze danych złożoną z naukowych potwierdzonych artykułów itp, a nie do ogólnie internetu ponieważ czesto wystepują błędne dane.

• odpowiadanie na pytania klientów w telefonicznej infolinii - może odpowiadać na czesto zadawana pytania do których została przygotowana przez twórców, może mówić zrozumiale.

• odpowiadanie na pytania klientów w internetowej infolinii - może odpowiadać na czesto zadawana pytania do których zostałą przygotowana przez twórców, dosć często odsyła do ludzkiego konsultanta ponieważ nie zna odpowiedzi.
