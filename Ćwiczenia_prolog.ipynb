##Ćwiczenie_1_WSSI
**1.1**

A. Są dla siebie rodzeństwem. Relacja ta zachodzi gdy X i Y: nie są tą samą osobą, mają tą samą matkę i tego sameo ojca.

B. Są dla siebie kuzynami. Relacja ta zachodzi gdy: rodzic X i rodzic Y są rodzeństwem.

C. Są dla siebie swatami. X ma potomka, który ma dziecko, którego drugi rodzic ma Y jako rodzica. X i Y są dziadkami tego samego dziecka. Relacja ta zachodzi gdy: X ma dziecko, Y ma dziecko, te dzieci razem mają dziecko, X nie jest tą samą osobą co Y, ich dzieci nie są tą samą osobą bo wtedy małżeństwo jest uznawane za swatów.

D. Y jest rodzicem przyrodnim X. Relacja ta zachodzi gdy: matka ma 2 dzieci z czego jedno to X, ale te dzieci mają różnych ojców i ojciec tego drugiego dziecka to Y lub gdy ojciec ma 2 dzieci z czego jedno to X, ale te dzieci mają różne matki i matka tego drugiego dziecka to Y.

E. Są rodzeństwem przyrodnim. Relacja ta zachodzi gdy: X i Y mają tą samą matkę ale róznych ojców lub gdy mają tego samego ojca ale różne matki. Nie tzreba dodawać warunku, że X i Y to inne osoby, ponieważ zapewnia to warunek róznych ojców/matek.

F. Szwagier/szwagierka/bratowa. Relacja ta zachodzi gdy: Y i R mają razem dziecko, R i X są rodzeństwem, Y i R to inne osoby lub gdy to X i R mają razem dziecko, R i Y są rodzeństwem, X i R to inne osoby.

G. Są rodzeństwem, ale X jest też wujkiem/ciocią Y. Nie znam nazwy na takie pokrewieństwo więc nazwę to relacja_g. Relacja ta zachodzi gdy: X i Y są rodzeństwem, drugi rodzic X ma dziecko, które ma dziecko, które jest drugim rodzicem Y.

**1.2**
```prolog
rodzic(X, Y) :-
    ojciec(X, Y);
    matka(X, Y).

dziadek(X, V) :- ojciec(X, Z),
    ojciec(Z, V); matka(Z, V).

babcia(X, V):- matka(X, Z),
    ojciec(Z, V); matka(Z, V).

siostra(X, Y) :-
    kobieta(X),
    matka(Z, X),
    matka(Z, Y),
    ojciec(V, X),
    ojciec(V, Y),
    X \= Y.

rodzeństwo(X, Y) :-
    X \= Y,
    matka(Z, X),
    matka(Z, Y),
    ojciec(V, X),
    ojciec(V, Y).

kuzyni(X, Y) :-
    rodzic(M, Y),
    rodzic(O, X),
    rodzeństwo(M, O).

swaci(X, Y) :-
    rodzic(X, K),
    rodzic(Y, M),
    rodzic(K, S),
    rodzic(M, S),
    X\=Y,
    K\=M.

rodzic_przyrodni(Y, X) :-
    matka(M, X), matka(M,S), ojciec(Y, S), ojciec(O, X), Y\=O;
    ojciec(O, X), ojciec(O,S), matka(Y, S), matka(M, X), Y\=M.

rodzeństwo_przyrodnie(X, Y) :-
    matka(M, X), matka(M, Y), ojciec(Z, X), ojciec(V, Y), Z\=V;
    ojciec(O, X), ojciec(O, Y), matka(Z, X), matka(V, Y), Z\=V.

szwagier_szwagierka_bratowa(X, Y) :-      
    rodzic(Y, S), rodzic(R, S), rodzeństwo(X, R), Y\=R;
    rodzic(X, S), rodzic(R, S), rodzeństwo(Y, R), X\=R.

relacja_g(X, Y) :-
    rodzeństwo(X, Y),
    rodzic(Z, X),
    rodzic(V, X),
    rodzic(N, Y),
    rodzic(V, Y).

```
