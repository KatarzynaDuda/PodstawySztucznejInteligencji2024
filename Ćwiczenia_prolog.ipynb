**Zadanie**: Zdefiniuj regułę `rodzic(X, Y)`, która określa, czy `X` jest rodzicem `Y`.

**Przykładowe dane i oczekiwany wynik**:

```prolog
dziadek(X, Y) :-
    osoba(X),
    mężczyzna(X),
    ojciec(X, Z),
    (ojciec(Z, Y); matka(Z, Y)).

 ?- dziadek(aleks, tosia).
 true

 ?- dziadek(karol, michał).
 true

 ?- dziadek(weronika, michał).
 false % bo to babcia
```
